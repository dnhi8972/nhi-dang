```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras import Input
import numpy as np

# ==========================================
# 1. TIỀN XỬ LÝ DỮ LIỆU HOA
# ==========================================
# Đổi đường dẫn tới thư mục chứa 5 loại hoa của bạn trên Drive
train_dir = "/content/drive/MyDrive/Đặng Ngọc Yến Nhi/hoa lá hẹ"

img_width, img_height = 200, 200
batch_size = 64

train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",
    validation_split=0.2 # Vẫn chia 80% train, 20% validation
)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='validation'
)

class_names = list(train_generator.class_indices.keys())
print("5 loại hoa hệ thống nhận diện được là:", class_names)

# ==========================================
# 2. XÂY DỰNG MÔ HÌNH CNN
# ==========================================
model = Sequential([
    Input(shape=(img_height, img_width, 3)),

    Conv2D(32, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),

    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(5, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ==========================================
# 3. HUẤN LUYỆN
# ==========================================
epochs = 40
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs
)

# Lưu lại mô hình hoa vào Drive
model.save('/content/drive/MyDrive/flower_recognition_cnn_50epochs.keras')
print("Đã hoàn thành train mô hình hoa và lưu thành công!")
```

```python
import numpy as np
from keras.utils import load_img, img_to_array
import matplotlib.pyplot as plt
from keras.models import load_model # Import load_model

# 1. Đường dẫn tới bức ảnh bông hoa bạn muốn test thử
test_image_path = "/content/drive/MyDrive/Colab Notebooks/hoa lá hẹ/hoa cẩm chướng/OIP (100).jpg"

# Load the trained model
model = load_model('/content/drive/MyDrive/flower_recognition_cnn_50epochs.keras')

class_names = ['hoa cẩm chướng', 'hoa cẩm tú cầu', 'hoa ly', 'hoa màu hồng', 'hoa tulip']
class_indices = {name: i for i, name in enumerate(class_names)}
labels_map = {v: k for k, v in class_indices.items()}

# 2. Đọc và xử lý ảnh 200x200 giống lúc train
img = load_img(test_image_path, target_size=(200, 200))
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# 3. Dự đoán loại hoa
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions)
confidence = np.max(predictions) * 100

# ÁP DỤNG BẢNG TRA CỨU ĐẢO NGƯỢC ĐỂ CHỐNG SAI NHÃN
predicted_flower = labels_map[predicted_class_index]

# 4. Hiển thị bông hoa và kết quả dự đoán
plt.imshow(img)
plt.axis('off')
plt.title(f"Kết quả: {predicted_flower} ({confidence:.2f}%)")
plt.show()
```
