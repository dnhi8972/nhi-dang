```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras import Input
import numpy as np

# ==========================================
# 1. TIỀN XỬ LÝ DỮ LIỆU (Giữ nguyên của bạn)
# ==========================================
train_dir = "/content/drive/MyDrive/Đặng Ngọc Yến Nhi/Ảnh selfie cho chiều t2 18 5"

img_width, img_height = 200, 200
batch_size = 64

train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='validation'
)

class_names = list(train_generator.class_indices.keys())

# ==========================================
# 2. XÂY DỰNG MÔ HÌNH CNN
# ==========================================
model = Sequential([
    Input(shape=(img_height, img_width, 3)),

    Conv2D(32, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),

    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(38, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ==========================================
# 3. HUẤN LUYỆN (Cày đủ 50 Epochs thẳng tiến)
# ==========================================
epochs = 50

# Đã bỏ tham số callbacks=[early_stop], máy sẽ chạy liên tục không dừng
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs
)

# Lưu lại mô hình vào Drive sau khi hoàn thành 50 epochs
model.save('/content/drive/MyDrive/face_recognition_cnn_50epochs.keras')
print("Đã hoàn thành 50 epochs và lưu mô hình thành công!")
```

```python
import numpy as np
from keras.models import load_model
from keras.utils import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# ==========================================
# 1. TẢI MÔ HÌNH ĐÃ LƯU TỪ DRIVE
# ==========================================
# Bạn kiểm tra xem tên file trên Drive của bạn là gì thì sửa lại cho đúng nhé
model_path = '/content/drive/MyDrive/face_recognition_cnn_50epochs.keras'
model = load_model(model_path)
print("Đã tải mô hình thành công!")

# ==========================================
# 2. TẠO TỰ ĐỘNG BẢNG TRA CỨU TÊN
# ==========================================
# Đường dẫn tới thư mục chứa ảnh gốc của bạn trên Drive
train_dir = "/content/drive/MyDrive/Colab Notebooks/Ảnh selfie cho chiều t2 18 5"

# Đoạn này chỉ để quét qua tên các thư mục con lấy nhãn (chạy siêu nhanh)
train_datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)
train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(200, 200), batch_size=32, class_mode='categorical', subset='training'
)

# Tạo bảng tra cứu ngược tự động (Từ Số lớp -> Tên người)
labels_map = {v: k for k, v in train_generator.class_indices.items()}

# ==========================================
# 3. TIẾN HÀNH NHẬN DIỆN ẢNH MỚI
# ==========================================
# Thay bằng đường dẫn tới bức ảnh khuôn mặt bạn muốn test thử
test_image_path = "/content/drive/MyDrive/Colab Notebooks/Ảnh test face/Anh2.jpeg"

# Đọc ảnh và xử lý chuẩn kích thước 200x200, chia 255 chuẩn học thuật
img = load_img(test_image_path, target_size=(200, 200))
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# Dự đoán khuôn mặt
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions)
confidence = np.max(predictions) * 100

# Tra cứu ra tên người
predicted_person = labels_map[predicted_class_index]

# ==========================================
# 4. HIỂN THỊ KẾT QUẢ
# ==========================================
plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.axis('off')
plt.title(f"Kết quả nhận diện: {predicted_person}\nĐộ tự tin: {confidence:.2f}%",
          fontsize=12, color='darkgreen', fontweight='bold')
plt.show()
```
