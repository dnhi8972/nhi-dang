```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras import Input
import numpy as np

# ==========================================
# 1. TIỀN XỬ LÝ DỮ LIỆU
# ==========================================
train_dir = "/content/drive/MyDrive/Đặng Ngọc Yến Nhi/Ảnh selfie cho chiều t2 18 5"

img_width, img_height = 200, 200
batch_size = 64

train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='validation'
)

class_names = list(train_generator.class_indices.keys())

# ==========================================
# 2. XÂY DỰNG MÔ HÌNH CNN
# ==========================================
model = Sequential([
    Input(shape=(img_height, img_width, 3)),

    Conv2D(32, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),

    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(38, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ==========================================
# 3. HUẤN LUYỆN
# ==========================================
epochs = 50

# Đã bỏ tham số callbacks=[early_stop], máy sẽ chạy liên tục không dừng
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs
)

# Lưu lại mô hình vào Drive sau khi hoàn thành 50 epochs
model.save('/content/drive/MyDrive/face_recognition_cnn_50epochs.keras')
print("Đã hoàn thành 50 epochs và lưu mô hình thành công!")
```

```python
import numpy as np
from keras.utils import load_img, img_to_array
import matplotlib.pyplot as plt
import os

# 1. Đường dẫn ảnh cần test
test_image_path = "/content/drive/MyDrive/Đặng Ngọc Yến Nhi/Ảnh selfie cho chiều t2 18 5/Duong Khanh Nghi/resized_z7856721749113_34e1e1e4438d0f3c72f024709a16d689.jpg"
train_dir = "/content/drive/MyDrive/Đặng Ngọc Yến Nhi/Ảnh selfie cho chiều t2 18 5"

# 2. Đọc và xử lý ảnh
img = load_img(test_image_path, target_size=(200, 200))
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# 3. Dự đoán
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions)
confidence = np.max(predictions) * 100

# 4. TỰ ĐỘNG LẤY DANH SÁCH TÊN TỪ THƯ MỤC
try:
    # Lấy danh sách tên thư mục con, sắp xếp theo thứ tự bảng chữ cái (giống cách Keras làm)
    folder_names = sorted([f for f in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, f))])
    predicted_person = folder_names[predicted_class_index]
except Exception as e:
    predicted_person = f"Lớp {predicted_class_index} (Lỗi lấy tên: {e})"

# 5. Hiển thị kết quả
plt.imshow(img)
plt.axis('off')
plt.title(f"Kết quả: {predicted_person} ({confidence:.2f}%)")
plt.show()
```
