```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras import Input
import numpy as np

# ==========================================
# 1. TIỀN XỬ LÝ DỮ LIỆU ĐƯỜNG CHỈ TAY
# ==========================================
# Đổi đường dẫn tới thư mục lớn chứa 2 thư mục con của bạn trên Drive
train_dir = "/content/drive/MyDrive/Colab Notebooks/Chỉ tayyyyy"

img_width, img_height = 200, 200
batch_size = 64

train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='validation'
)

class_names = list(train_generator.class_indices.keys())
print("2 lớp nhận diện bao gồm:", class_names)

# ==========================================
# 2. XÂY DỰNG MÔ HÌNH CNN
# ==========================================
model = Sequential([
    Input(shape=(img_height, img_width, 3)),

    Conv2D(32, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),

    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ==========================================
# 3. HUẤN LUYỆN
# ==========================================
epochs = 40
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs
)

# Lưu lại mô hình chỉ tay vào Drive
model.save('/content/drive/MyDrive/palm_recognition_cnn_50epochs.keras')
print("Đã hoàn thành train mô hình chỉ tay và lưu thành công!")
```

```python
import numpy as np
from keras.utils import load_img, img_to_array
from keras.models import load_model
import matplotlib.pyplot as plt

# 1. Đường dẫn ảnh test
test_image_path = "/content/drive/MyDrive/Colab Notebooks/Chỉ tayyyyy/Tay bình thường/IMG_9610.jpg"

# 2. Tải mô hình tốt nhất đã được bộ Checkpoint lưu lại
model = load_model('/content/drive/MyDrive/palm_perfect_model.keras')

# 3. Xử lý ảnh đầu vào
img = load_img(test_image_path, target_size=(200, 200))
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# 4. Dự đoán kết quả
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions)
confidence = np.max(predictions) * 100

# Tra cứu nhãn (Nếu chạy ở phiên mới, bạn có thể thay bằng danh sách thủ công)
# labels_map = {0: "Chữ M", 1: "Tay bình thường"}
labels_map = {v: k for k, v in train_generator.class_indices.items()}
predicted_palm = labels_map[predicted_class_index]

# Thêm lời luận giải tâm linh ngắn
if "M" in predicted_palm.upper():
    loi_boi_toan = "Số mệnh giàu sang, trực giác nhạy bén, hậu vận vô cùng rực rỡ!"
else:
    loi_boi_toan = "Tính cách thực tế, kiên cường, cuộc sống bình yên và gia đình hạnh phúc!"

# 5. Hiển thị kết quả hoàn chỉnh
plt.imshow(img)
plt.axis('off')
plt.title(f"Kết quả: {predicted_palm} ({confidence:.2f}%)\n\n🔮 Luận giải: {loi_boi_toan}",
          fontsize=12, color='darkblue', wrap=True)
plt.show()
```
