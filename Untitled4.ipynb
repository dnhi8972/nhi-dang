```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl

# 1. THIẾT LẬP CÁC BIẾN ĐẦU VÀO (Antecedents) VÀ ĐẦU RA (Consequents)
distance = ctrl.Antecedent(np.arange(0, 51, 1), 'distance')    # 0-50 km
traffic = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic')      # 0-100%
demand = ctrl.Antecedent(np.arange(0, 101, 1), 'demand')        # 0-100%
weather = ctrl.Antecedent(np.arange(0, 11, 1), 'weather')       # 0-10 (Tốt -> Xấu)

price = ctrl.Consequent(np.arange(0, 101, 1), 'price')          # Mức giá (thang điểm 100)
bonus = ctrl.Consequent(np.arange(0, 101, 1), 'bonus')          # Điểm thưởng (thang điểm 100)

# 2. ĐỊNH NGHĨA HÀM THÀNH VIÊN
distance['ngan'] = fuzzy.trimf(distance.universe, [0, 0, 8])
distance['trung_binh'] = fuzzy.trimf(distance.universe, [2, 8, 20])
distance['xa'] = fuzzy.trimf(distance.universe, [6, 20, 50])
distance['rat_xa'] = fuzzy.trapmf(distance.universe, [15, 30, 50, 50])

traffic['thap'] = fuzzy.trimf(traffic.universe, [0, 0, 30])
traffic['trung_binh'] = fuzzy.trimf(traffic.universe, [20, 50, 70])
traffic['cao'] = fuzzy.trimf(traffic.universe, [60, 100, 100])

demand['thap'] = fuzzy.trimf(demand.universe, [0, 0, 30])
demand['trung_binh'] = fuzzy.trimf(demand.universe, [20, 50, 70])
demand['cao'] = fuzzy.trimf(demand.universe, [60, 100, 100])

weather['tot'] = fuzzy.trimf(weather.universe, [0, 0, 5])
weather['xau'] = fuzzy.trimf(weather.universe, [5, 10, 10])

# Đầu ra: Giá và Thưởng
price['thap'] = fuzzy.trimf(price.universe, [0, 0, 40])
price['trung_binh'] = fuzzy.trimf(price.universe, [30, 50, 70])
price['cao'] = fuzzy.trimf(price.universe, [60, 80, 100])
price['rat_cao'] = fuzzy.trimf(price.universe, [80, 100, 100])

bonus['khong'] = fuzzy.trimf(bonus.universe, [0, 0, 20])
bonus['it'] = fuzzy.trimf(bonus.universe, [10, 30, 50])
bonus['trung_binh'] = fuzzy.trimf(bonus.universe, [40, 60, 80])
bonus['cao'] = fuzzy.trimf(bonus.universe, [70, 100, 100])

# 3. HỆ THỐNG LUẬT
rules = [
    # Nhóm luật tính GIÁ (Ví dụ một số luật tiêu biểu từ 1-10)
    ctrl.Rule(distance['ngan'] & traffic['thap'] & demand['thap'], price['thap']),
    ctrl.Rule(distance['ngan'] & traffic['trung_binh'] & demand['cao'], price['trung_binh']),
    ctrl.Rule(distance['trung_binh'] & traffic['cao'] & demand['cao'], price['cao']),
    ctrl.Rule(distance['xa'] & weather['xau'], price['cao']),
    ctrl.Rule(distance['rat_xa'] & traffic['cao'], price['rat_cao']),

    # Nhóm luật tính THƯỞNG
    ctrl.Rule(distance['xa'] & traffic['cao'], bonus['cao']),
    ctrl.Rule(weather['xau'] & demand['cao'], bonus['cao']),
    ctrl.Rule(distance['ngan'] & traffic['thap'], bonus['khong'])
]

# 4. KHỞI TẠO MÔ PHỎNG
grab_system = ctrl.ControlSystem(rules)
grab_sim = ctrl.ControlSystemSimulation(grab_system)
# 5. NHẬP DỮ LIỆU THỰC TẾ ĐỂ KIỂM TRA
grab_sim.input['distance'] = 15      # Đi 15km
grab_sim.input['traffic'] = 85       # Kẹt xe nặng (85%)
grab_sim.input['demand'] = 90        # Nhu cầu cực cao (90%)
grab_sim.input['weather'] = 8        # Trời mưa to (mức 8/10)

# Tính toán
grab_sim.compute()

# Xuất kết quả
print(f"--- KẾT QUẢ TỪ HỆ THỐNG MỜ ---")
print(f"Mức giá đề xuất (thang 100): {grab_sim.output['price']:.2f}")
print(f"Mức thưởng tài xế (thang 100): {grab_sim.output['bonus']:.2f}")

# Vẽ đồ thị trực quan (chỉ chạy được trên Colab)
price.view(sim=grab_sim)
```

```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl

# 1. ĐỊNH NGHĨA CÁC BIẾN ĐẦU VÀO VÀ ĐẦU RA
# Universe: Các dải giá trị (ví dụ Rating từ 0-5 sao, Sales từ 0-100%)
rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'rating')
sales = ctrl.Antecedent(np.arange(0, 101, 1), 'sales')
margin = ctrl.Antecedent(np.arange(0, 101, 1), 'margin')
season = ctrl.Antecedent(np.arange(0, 11, 1), 'season')
competitor = ctrl.Antecedent(np.arange(0, 101, 1), 'competitor')

discount = ctrl.Consequent(np.arange(0, 71, 1), 'discount')

# 2. THIẾT LẬP HÀM THÀNH VIÊN (Membership Functions)
# Dựa trên mô tả: Thấp, Trung bình, Cao
rating['thap'] = fuzzy.trimf(rating.universe, [0, 0, 4.0])
rating['trung_binh'] = fuzzy.trimf(rating.universe, [3.5, 4.25, 4.7])
rating['cao'] = fuzzy.trimf(rating.universe, [4.5, 5, 5])

sales['thap'] = fuzzy.trimf(sales.universe, [0, 0, 40])
sales['trung_binh'] = fuzzy.trimf(sales.universe, [30, 50, 70])
sales['cao'] = fuzzy.trimf(sales.universe, [60, 100, 100])

margin['thap'] = fuzzy.trimf(margin.universe, [0, 0, 30])
margin['trung_binh'] = fuzzy.trimf(margin.universe, [25, 50, 75])
margin['cao'] = fuzzy.trimf(margin.universe, [70, 100, 100])

season['khong'] = fuzzy.trimf(season.universe, [0, 0, 3])
season['trung_binh'] = fuzzy.trimf(season.universe, [2, 5, 8])
season['cao'] = fuzzy.trimf(season.universe, [7, 10, 10])

competitor['thap'] = fuzzy.trimf(competitor.universe, [0, 0, 40])
competitor['trung_binh'] = fuzzy.trimf(competitor.universe, [30, 50, 70])
competitor['cao'] = fuzzy.trimf(competitor.universe, [60, 100, 100])

# Đầu ra: Tỷ lệ chiết khấu (%)
discount['rat_thap'] = fuzzy.trimf(discount.universe, [0, 2.5, 5])
discount['thap'] = fuzzy.trimf(discount.universe, [5, 7.5, 10])
discount['trung_binh'] = fuzzy.trimf(discount.universe, [10, 15, 20])
discount['cao'] = fuzzy.trimf(discount.universe, [20, 30, 40])
discount['rat_cao'] = fuzzy.trimf(discount.universe, [40, 55, 70])

# 3. HỆ THỐNG LUẬT MỜ
rule1 = ctrl.Rule(rating['cao'] & sales['cao'] & margin['cao'], discount['rat_thap'])
rule2 = ctrl.Rule(rating['thap'] & sales['thap'] & margin['cao'], discount['cao'])
rule3 = ctrl.Rule(season['cao'] & competitor['cao'], discount['rat_cao'])
rule4 = ctrl.Rule(rating['trung_binh'] & sales['trung_binh'] & margin['trung_binh'], discount['trung_binh'])
rule5 = ctrl.Rule(competitor['thap'] & margin['thap'] & sales['cao'], discount['rat_thap'])

# 4. TẠO HỆ THỐNG ĐIỀU KHIỂN
discount_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
discount_sim = ctrl.ControlSystemSimulation(discount_ctrl)

# 5. NHẬP DỮ LIỆU VÀ CHẠY THỬ NGHIỆM
# Giả sử: Shop 4.8 sao, Doanh số 80%, Lợi nhuận 80%, Đang mùa lễ hội (9), Đối thủ giảm giá mạnh (90%)
discount_sim.input['rating'] = 4.8
discount_sim.input['sales'] = 80
discount_sim.input['margin'] = 80
discount_sim.input['season'] = 9
discount_sim.input['competitor'] = 90

# Tính toán kết quả
discount_sim.compute()

print(f"Mức chiết khấu đề xuất: {discount_sim.output['discount']:.2f}%")
discount.view(sim=discount_sim)
```

```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl

# 1. Khai báo các biến đầu vào và đầu ra
demand = ctrl.Antecedent(np.arange(0, 101, 1), 'demand')
comp_pressure = ctrl.Antecedent(np.arange(0, 101, 1), 'comp_pressure')
reputation = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'reputation')
profit_margin = ctrl.Antecedent(np.arange(0, 101, 1), 'profit_margin')
seasonal = ctrl.Antecedent(np.arange(0, 11, 1), 'seasonal')

discount = ctrl.Consequent(np.arange(0, 71, 1), 'discount')

# 2. Định nghĩa hàm thành viên (Membership functions)
demand.automf(3, names=['low', 'medium', 'high'])
comp_pressure.automf(3, names=['low', 'medium', 'high'])
reputation.automf(3, names=['low', 'medium', 'high'])
profit_margin.automf(3, names=['low', 'medium', 'high'])
seasonal['none'] = fuzzy.trimf(seasonal.universe, [0, 0, 2])
seasonal['medium'] = fuzzy.trimf(seasonal.universe, [1, 5, 8])
seasonal['high'] = fuzzy.trimf(seasonal.universe, [7, 10, 10])

discount['very_low'] = fuzzy.trimf(discount.universe, [0, 2.5, 5])
discount['low'] = fuzzy.trimf(discount.universe, [5, 7.5, 10])
discount['medium'] = fuzzy.trimf(discount.universe, [10, 15, 25])
discount['high'] = fuzzy.trimf(discount.universe, [20, 30, 40])
discount['very_high'] = fuzzy.trimf(discount.universe, [40, 55, 70])

# 3. Hệ thống luật mờ
rules = [
    ctrl.Rule(demand['high'] & comp_pressure['low'] & profit_margin['low'], discount['very_low']),
    ctrl.Rule(demand['low'] & comp_pressure['high'] & profit_margin['high'], discount['high']),
    ctrl.Rule(reputation['high'] & profit_margin['medium'] & seasonal['high'], discount['medium']),
    ctrl.Rule(comp_pressure['high'] & seasonal['high'] & profit_margin['high'], discount['very_high']),
    ctrl.Rule(reputation['low'] & demand['medium'] & profit_margin['low'], discount['medium']),
    ctrl.Rule(demand['high'] & seasonal['none'] & comp_pressure['low'], discount['very_low']),
    ctrl.Rule(profit_margin['high'] & comp_pressure['medium'] & seasonal['high'], discount['medium'])
]

# 4. Thiết lập mô phỏng
shopee_ctrl = ctrl.ControlSystem(rules)
shopee_sim = ctrl.ControlSystemSimulation(shopee_ctrl)

# 5. Tình huống thực tế (Trang 91)
shopee_sim.input['demand'] = 85           # Cao
shopee_sim.input['comp_pressure'] = 50    # Trung bình
shopee_sim.input['reputation'] = 4.2      # Trung bình
shopee_sim.input['profit_margin'] = 80    # Cao
shopee_sim.input['seasonal'] = 9          # Cao (Sự kiện 11.11)

shopee_sim.compute()

print("=== KẾT QUẢ BÀI 2.13 ===")
print(f"Mức chiết khấu đề xuất: {shopee_sim.output['discount']:.2f}%")
discount.view(sim=shopee_sim)
```

```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl

# 1. Khai báo các biến đầu vào và đầu ra
density = ctrl.Antecedent(np.arange(0, 101, 1), 'density')
urgency = ctrl.Antecedent(np.arange(0, 101, 1), 'urgency')
load = ctrl.Antecedent(np.arange(0, 101, 1), 'load')
traffic = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic')
profit = ctrl.Antecedent(np.arange(0, 101, 1), 'profit')

combine = ctrl.Consequent(np.arange(0, 11, 1), 'combine') # Số lượng đơn kết hợp
priority = ctrl.Consequent(np.arange(0, 101, 1), 'priority') # Độ ưu tiên

# 2. Định nghĩa hàm thành viên
for var in [density, urgency, load, traffic, profit]:
    var.automf(3, names=['low', 'medium', 'high'])

combine['few'] = fuzzy.trimf(combine.universe, [0, 0, 4])
combine['some'] = fuzzy.trimf(combine.universe, [3, 5, 7])
combine['many'] = fuzzy.trimf(combine.universe, [6, 10, 10])

priority['low'] = fuzzy.trimf(priority.universe, [0, 0, 40])
priority['medium'] = fuzzy.trimf(priority.universe, [30, 50, 70])
priority['high'] = fuzzy.trimf(priority.universe, [60, 100, 100])

# 3. Hệ thống luật (8 luật - Trang 93)
logistics_rules = [
    # Luật kết hợp đơn hàng
    ctrl.Rule(density['high'] & load['low'] & traffic['medium'], combine['some']), # Modified rule to ensure it fires
    ctrl.Rule(density['medium'] & traffic['high'] & urgency['medium'], combine['some']),
    ctrl.Rule(load['high'] & density['high'] & profit['medium'], combine['some']),
    ctrl.Rule(density['low'] & urgency['high'] & traffic['medium'], combine['some']),
    ctrl.Rule(profit['high'] & urgency['high'] & traffic['high'], combine['some']),
    # Luật ưu tiên giao hàng
    ctrl.Rule(urgency['high'] & profit['high'], priority['high']),
    ctrl.Rule(urgency['medium'] & traffic['medium'], priority['medium']),
    ctrl.Rule(urgency['low'] & density['high'] & profit['low'], priority['low'])
]

# 4. Thiết lập mô phỏng
logistics_ctrl = ctrl.ControlSystem(logistics_rules)
logistics_sim = ctrl.ControlSystemSimulation(logistics_ctrl)

# 5. Dữ liệu đầu vào thực tế (Trang 93)
logistics_sim.input['density'] = 85    # Cao
logistics_sim.input['urgency'] = 50    # Trung bình
logistics_sim.input['load'] = 20       # Thấp
logistics_sim.input['traffic'] = 50    # Trung bình
logistics_sim.input['profit'] = 50     # Trung bình

logistics_sim.compute()

print("=== KẾT QUẢ BÀI 2.14 ===")
print(f"Số đơn hàng nên kết hợp: {logistics_sim.output['combine']:.1f} đơn")
print(f"Mức độ ưu tiên giao hàng: {logistics_sim.output['priority']:.2f}/100")
```
