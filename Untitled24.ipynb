```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

# 1. ĐIỀN ĐƯỜNG DẪN ĐẾN THƯ MỤC DATASET MỚI CỦA BẠN TRÊN DRIVE
PATH_TRAIN = '/content/drive/MyDrive/CNN project/train'       # Sửa lại cho đúng cấu trúc Drive của bạn
PATH_VALID = '/content/drive/MyDrive/CNN project/validation'  # Sửa lại cho đúng cấu trúc Drive của bạn

# 2. Cấu hình tăng cường ảnh cho tập Train (Giúp mô hình chống học vẹt)
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input, # Ép màu chuẩn MobileNetV2
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Tập Validation chỉ cần chuẩn hóa màu, không cần làm biến dạng ảnh
valid_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

# 3. Tạo bộ nạp ảnh tự động (Target size bắt buộc là 224x224)
print("📸 Đang nạp dữ liệu tập TRAIN:")
train_generator = train_datagen.flow_from_directory(
    PATH_TRAIN,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

print("\n📸 Đang nạp dữ liệu tập VALIDATION:")
validation_generator = valid_datagen.flow_from_directory(
    PATH_VALID,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

# In kiểm tra danh sách 12 món ăn để đảm bảo Colab đã nhận diện đủ
print("\n✅ Danh sách món ăn đã nhận diện thành công:")
print(train_generator.class_indices)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout

# 1. Lấy bộ não MobileNetV2 đã học sẵn từ ImageNet
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Đóng băng các lớp cũ của Google để giữ lại kiến thức nền tảng
for layer in base_model.layers:
    layer.trainable = False

# 2. Thiết kế thêm lớp Output mới fit chuẩn với 13 món ăn của bạn
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.4)(x)  # Giúp mô hình không bị overfitting (học vẹt)


predictions = Dense(12, activation='softmax')(x)

# 3. Lắp ráp và biên dịch mô hình
model_transfer = Model(inputs=base_model.input, outputs=predictions)
model_transfer.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model_transfer.summary() # In bảng tổng quan kiến trúc mạng ra xem
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# ĐƯỜNG DẪN MỚI TOÀN DIỆN ĐỂ LƯU MÔ HÌNH CHUẨN ĐÉT
NEW_CHECKPOINT_PATH = '/content/drive/MyDrive/Do_An_YOLO/mo_hinh_mobilenet_12_mon_HOAN_HAO.h5'

# Cấu hình tự động lưu lại phiên bản có val_accuracy cao kỷ lục
checkpoint = ModelCheckpoint(
    filepath=NEW_CHECKPOINT_PATH,
    monitor='val_accuracy',   # Theo dõi chặt chẽ độ chính xác tập Validation thực tế
    mode='max',
    save_best_only=True,      # Chỉ ghi đè khi mô hình thông minh hơn phiên bản cũ
    verbose=1
)

# Cấu hình tự động dừng nếu sau 5 vòng học liên tiếp mà val_accuracy không tăng (Chống học vẹt)
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=5,
    restore_best_weights=True, # Tự động hồi phục lại trọng số ở vòng tốt nhất sau khi dừng
    verbose=1
)

print("🚀 BẮT ĐẦU CHẠY HUẤN LUYỆN LẠI TỪ ĐẦU (SẠCH SẼ & CHUẨN XÁC)...")

# Tiến hành bấm nút Train
history = model_transfer.fit(
    train_generator,
    epochs=30,                           # Cho phép học sâu tối đa 30 vòng
    validation_data=validation_generator, # Đối chiếu trực tiếp với tập Validation chuẩn tên
    callbacks=[checkpoint, early_stopping] # Kích hoạt bộ đôi bảo vệ thông minh
)
```
